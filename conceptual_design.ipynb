```python
def LD_max_approx(Sratio, K, AR ):
    return K*math.sqrt(AR/Sratio)

def effective_AR(b, h, S):
    return b**2 / S * (1 + 1.9*h/b)
```

```python
Sratio = 2.1
K = 11
b = 1.8
h = 0.1
r_avg = 0.3
S = b*r_avg
AR = effective_AR(b, h, S)
LD_max = LD_max_approx(Sratio, K, AR)
print(f"AR: {b*b/S:.2f}"
      f"\nEffective AR: {AR:.2f}"
      f"\nEstimated L/D max: {LD_max:.2f}")
```
